```python
!pip install nltk
import nltk
from nltk.chat.util import Chat, reflections

# הורדת חבילות הנתונים הרלוונטיות לעיבוד שפה טבעית
nltk.download('punkt')
nltk.download('wordnet')
```

```python
# הגדרת התבניות והתשובות בהתאם למערכת זיהוי המחלות
patterns = [
    (r'hi|hello|hey',
     ['Hello! Welcome to the Tomato Disease Search Engine. How can I help you?',
      'Hi there! Looking for organic farming solutions today?']),

    (r'how does the (.*) work\?',
     ['Our search engine uses image recognition to identify diseases in tomato crops. Just upload an image of the affected plant!']),

    (r'what is (.*)\?',
     ['Our index contains various agricultural data about %1. Try uploading a photo of it so our model can provide a precise diagnosis.']),

    (r'(.*) organic farming(.*)',
     ['For organic farming, we prioritize natural treatments. Once our system identifies the disease, it will suggest strictly organic solutions.']),

    (r'(.*) identify (.*)',
     ['To identify %2, please capture a clear image of the tomato leaves and upload it to the system.']),

    (r'quit|bye|exit',
     ['Goodbye! Wishing you a healthy harvest!'])
]

# יצירת אובייקט הצ'אטבוט
chatbot = Chat(patterns, reflections)
```

```python
# התחלת השיחה
print("Welcome to the Tomato AI Assistant. Type 'quit' to exit.")

while True:
    user_input = input("You: ")

    # תנאי יציאה מהלולאה
    if user_input.lower() in ['quit', 'bye', 'exit']:
        print("Chatbot: Goodbye!")
        break

    response = chatbot.respond(user_input)

    if response:
        print("Chatbot:", response)
    else:
        print("Chatbot: I am not sure about that. Try asking about tomato diseases, image recognition, or organic treatments.")
```
